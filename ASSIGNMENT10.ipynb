```
'''Q1. Write a unique paragraph (5-6 sentences) about your favorite topic (e.g., sports,
technology, food, books, etc.).
1. Convert text to lowercase and remove punctuaƟon using re.
2. Tokenize the text into words and sentences.
3. Split using split() and word_tokenize() and compare how Python split and NLTK’s
word_tokenize() differ.
4. Remove stopwords (using NLTK's stopwords list).
5. Display word frequency distribuƟon (excluding stopwords).
'''




# Install necessary libraries
!pip install nltk --quiet

# Import necessary libraries
import nltk
import re
from nltk.tokenize import word_tokenize, sent_tokenize
from nltk.corpus import stopwords
from nltk.probability import FreqDist

# Download required NLTK resources explicitly
nltk.download('punkt')       # For sentence and word tokenization
nltk.download('stopwords')   # For stopwords removal
nltk.download('punkt_tab')   # Explicitly download punkt_tab, as it's missing

# Step 1: Write a unique paragraph (favorite topic)
paragraph = """Books have always been my sanctuary from the noise of everyday life.
They allow me to travel to different worlds without leaving my room.
From fantasy lands filled with dragons to realistic tales about human struggles,
books offer endless possibilities. I enjoy collecting both fiction and nonfiction,
and every shelf in my room is filled with stories waiting to be explored.
Reading not only improves my imagination but also expands my knowledge in unexpected ways."""

# Step 2: Convert text to lowercase and remove punctuation using regex
paragraph_lower = paragraph.lower()  # Convert to lowercase
paragraph_clean = re.sub(r'[^\w\s]', '', paragraph_lower)  # Remove punctuation

# Step 3: Tokenize the text into words and sentences
sentences = sent_tokenize(paragraph)  # Sentence tokenization
words_split = paragraph_clean.split()  # Using split() (basic Python)
words_tokenized = word_tokenize(paragraph_clean)  # Using NLTK's word_tokenize()

# Step 4: Remove stopwords using NLTK's stopwords list
stop_words = set(stopwords.words('english'))  # Get list of stopwords
filtered_words = [word for word in words_tokenized if word not in stop_words]  # Remove stopwords

# Step 5: Display word frequency distribution (excluding stopwords)
fdist = FreqDist(filtered_words)  # Frequency distribution of filtered words

# Results
print("\nOriginal Paragraph:\n", paragraph)
print("\nLowercased and punctuation removed:\n", paragraph_clean)
print("\nTokenized Sentences:", sentences)
print("\nWords using split():", words_split)
print("\nWords using word_tokenize():", words_tokenized)
print("\nWords after removing stopwords:", filtered_words)

# Display word frequency distribution (excluding stopwords)
print("\nWord Frequency Distribution (excluding stopwords):")
for word, freq in fdist.most_common():
    print(f"{word}: {freq}")
```

```
'''Q2. Using the same paragraph from Q1:
1. Extract all words with only alphabets using re.findall()
2. Remove stop words using NLTK’s stopword list
3. Perform stemming with PorterStemmer
4. Perform lemmaƟzaƟon with WordNetLemmaƟzer
5. Compare the stemmed and lemmaƟzed outputs and explain when you’d prefer one over
the other. '''
# Install necessary libraries
!pip install nltk --quiet

# Import necessary libraries
import nltk
import re
from nltk.tokenize import word_tokenize
from nltk.corpus import stopwords
from nltk.stem import PorterStemmer, WordNetLemmatizer

# Download required NLTK resources explicitly
nltk.download('punkt')       # For sentence and word tokenization
nltk.download('stopwords')   # For stopwords removal
nltk.download('wordnet')     # For lemmatization
nltk.download('omw-1.4')     # For wordnet support (helps with lemmatization)

# Step 1: Write a unique paragraph (favorite topic)
paragraph = """Books have always been my sanctuary from the noise of everyday life.
They allow me to travel to different worlds without leaving my room.
From fantasy lands filled with dragons to realistic tales about human struggles,
books offer endless possibilities. I enjoy collecting both fiction and nonfiction,
and every shelf in my room is filled with stories waiting to be explored.
Reading not only improves my imagination but also expands my knowledge in unexpected ways."""

# Step 2: Extract all words with only alphabets using re.findall()
words_alphabets_only = re.findall(r'\b[a-zA-Z]+\b', paragraph.lower())  # Extract only alphabetic words
print("\nWords with only alphabets:", words_alphabets_only)

# Step 3: Remove stopwords using NLTK's stopword list
stop_words = set(stopwords.words('english'))  # Get list of stopwords
filtered_words = [word for word in words_alphabets_only if word not in stop_words]  # Remove stopwords
print("\nFiltered words after stopword removal:", filtered_words)

# Step 4: Perform stemming with PorterStemmer
porter_stemmer = PorterStemmer()
stemmed_words = [porter_stemmer.stem(word) for word in filtered_words]  # Apply stemming
print("\nStemmed words using PorterStemmer:", stemmed_words)

# Step 5: Perform lemmatization with WordNetLemmatizer
lemmatizer = WordNetLemmatizer()
lemmatized_words = [lemmatizer.lemmatize(word) for word in filtered_words]  # Apply lemmatization
print("\nLemmatized words using WordNetLemmatizer:", lemmatized_words)

# Step 6: Compare the stemmed and lemmatized outputs and explain when you'd prefer one over the other
print("\nComparison:")
print(f"Stemmed Words: {stemmed_words}")
print(f"Lemmatized Words: {lemmatized_words}")

# Explanation of when you'd prefer stemming or lemmatization:
explanation = """
- **Stemming** is a quicker, rule-based process that cuts off the suffixes of words to reduce them to a root form. It might result in words that aren't actual words, which can be problematic in some cases. Use stemming when speed is more important than precision.

- **Lemmatization** involves converting words into their base or dictionary form (lemmas), which is more accurate. The process is slower than stemming but produces more meaningful and correct word forms. Use lemmatization when you need accurate word representation for tasks like text classification, sentiment analysis, etc.
"""
print(explanation)
```

```
'''Q3. Choose 3 short texts of your own (e.g., different news headlines, product reviews).
1. Use CountVectorizer to generate the Bag of Words representaƟon.
2. Use TfidfVectorizer to compute TF-IDF scores.
3. Print and interpret the top 3 keywords from each text using TF-IDF. '''

# Install necessary libraries
!pip install scikit-learn --quiet

# Import necessary libraries
from sklearn.feature_extraction.text import CountVectorizer, TfidfVectorizer
import numpy as np

# Sample texts (You can replace these with your own texts)
texts = [
    "The weather is sunny today with clear skies and a light breeze.",
    "The new phone model has a powerful camera and long battery life.",
    "Stock market prices have been fluctuating this week due to economic uncertainty."
]

# Step 1: Using CountVectorizer to generate the Bag of Words representation
count_vectorizer = CountVectorizer()
X_count = count_vectorizer.fit_transform(texts)  # This will create the Bag of Words matrix

# Step 2: Using TfidfVectorizer to compute TF-IDF scores
tfidf_vectorizer = TfidfVectorizer()
X_tfidf = tfidf_vectorizer.fit_transform(texts)  # This will compute the TF-IDF matrix

# Step 3: Print the Bag of Words
print("\nBag of Words Representation:")
words = count_vectorizer.get_feature_names_out()
print(words)

# Step 4: Print the TF-IDF scores
print("\nTF-IDF Scores:")
tfidf_scores = X_tfidf.toarray()
for i, text in enumerate(texts):
    print(f"\nText {i+1}: {text}")
    for j in range(len(words)):
        print(f"{words[j]}: {tfidf_scores[i][j]:.4f}")

# Step 5: Interpret the top 3 keywords based on TF-IDF scores
top_keywords = []
for i, text in enumerate(texts):
    tfidf_scores_for_text = tfidf_scores[i]
    sorted_indices = np.argsort(tfidf_scores_for_text)[::-1]  # Sort indices by TF-IDF score in descending order
    top_3_indices = sorted_indices[:3]  # Get the top 3 indices

    top_keywords_for_text = [words[index] for index in top_3_indices]
    top_keywords.append(top_keywords_for_text)

    print(f"\nTop 3 keywords for Text {i+1}: {', '.join(top_keywords_for_text)}")
```

```
'''Q4. Write 2 short texts (4–6 lines each) describing two different technologies (e.g., AI vs
Blockchain).
1. Preprocess and tokenize both texts.
2. Calculate:
a. Jaccard Similarity using sets
b. Cosine Similarity using TfidfVectorizer + cosine_similarity()
c. Analyze which similarity metric gives beƩer insights in your case.'''

# Install necessary libraries
!pip install nltk scikit-learn --quiet

# Import modules
import nltk
from nltk.tokenize import word_tokenize
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity

# Download NLTK data
nltk.download('punkt')

# --- Q4 Starts ---
print("\n--- Q4: Similarity Measures ---\n")

# Texts
text1 = """Artificial Intelligence is transforming industries with machine learning algorithms that mimic human intelligence."""
text2 = """Blockchain is revolutionizing industries by enabling secure and transparent transactions without intermediaries."""
nltk.download('punkt_tab')
# Preprocess and tokenize
tokens1 = [word for word in word_tokenize(text1.lower()) if word.isalpha()]
tokens2 = [word for word in word_tokenize(text2.lower()) if word.isalpha()]

# Jaccard Similarity
set1 = set(tokens1)
set2 = set(tokens2)
jaccard_similarity = len(set1.intersection(set2)) / len(set1.union(set2))
print("Jaccard Similarity:", jaccard_similarity)

# Cosine Similarity
texts = [text1, text2]
tfidf_vectorizer = TfidfVectorizer()
tfidf_matrix = tfidf_vectorizer.fit_transform(texts)
cosine_sim = cosine_similarity(tfidf_matrix[0:1], tfidf_matrix[1:2])
print("Cosine Similarity:", cosine_sim[0][0])

# Analysis
print("\nAnalysis:")
print("Jaccard similarity measures common words ignoring frequency.")
print("Cosine similarity considers word importance and is generally better for longer texts.")
```

```
'''Q6. Choose your own paragraph (~100 words) as training data.
1. Tokenize text using Tokenizer() from keras.preprocessing.text
2. Create input sequences and build a simple LSTM or Dense model
3. Train the model and generate 2–3 new lines of text starƟng from any seed word you
provide. '''
# Install TensorFlow
!pip install tensorflow --quiet

# Import required libraries
import tensorflow as tf
from tensorflow.keras.preprocessing.text import Tokenizer
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import LSTM, Dense, Embedding
from tensorflow.keras.preprocessing.sequence import pad_sequences
import numpy as np

# Step 1: Choose your own paragraph (Training Data)
paragraph = """Artificial Intelligence is the simulation of human intelligence processes by machines, especially computer systems.
These processes include learning, reasoning, and self-correction. AI is becoming an important part of various industries such as healthcare,
finance, and automotive. It has the potential to improve the efficiency of processes and bring innovative solutions to complex problems.
With advancements in machine learning and deep learning, AI is expected to continue growing and changing the way we live and work."""

# Step 2: Tokenize text using Tokenizer
tokenizer = Tokenizer()
tokenizer.fit_on_texts([paragraph])

# Convert the text to sequences of integers
sequences = tokenizer.texts_to_sequences([paragraph])[0]

# Prepare the sequences for training: Use input sequences (X) and next words as targets (y)
X, y = [], []
sequence_length = 5  # Number of words in each sequence (input)

for i in range(sequence_length, len(sequences)):
    X.append(sequences[i-sequence_length:i])
    y.append(sequences[i])

# Convert X and y into numpy arrays
X = np.array(X)
y = np.array(y)

# Step 3: Build the LSTM model
model = Sequential()
model.add(Embedding(input_dim=len(tokenizer.word_index) + 1, output_dim=10, input_length=sequence_length))
model.add(LSTM(50))
model.add(Dense(len(tokenizer.word_index) + 1, activation='softmax'))

# Compile the model
model.compile(loss='sparse_categorical_crossentropy', optimizer='adam', metrics=['accuracy'])

# Step 4: Train the model
model.fit(X, y, epochs=500, verbose=0)

# Step 5: Generate text starting from a seed word
def generate_text(seed_text, num_words):
    for _ in range(num_words):
        # Convert seed text into sequence of integers
        sequence = tokenizer.texts_to_sequences([seed_text])[0]

        # Pad the sequence to the required length (sequence_length)
        sequence = pad_sequences([sequence], maxlen=sequence_length, padding='pre')

        # Predict the next word
        predicted = model.predict(sequence, verbose=0)

        # Get the index of the predicted word
        predicted_word_index = np.argmax(predicted)

        # Get the word from the index
        predicted_word = tokenizer.index_word[predicted_word_index]

        # Append the predicted word to the seed text
        seed_text += " " + predicted_word
    return seed_text

# Example seed word to generate text
seed_word = "artificial"
generated_text = generate_text(seed_word, num_words=10)  # Generate 10 words starting from "artificial"

print(generated_text)
```
