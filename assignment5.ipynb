```python
import numpy as np
# 1 For the array gfg = np.matrix('[4, 1, 9; 12, 3, 1; 4, 5, 6]'), Find
gfg=np.matrix('[4,1,9;12,3,1;4,5,6]')
#i. Sum of all elements
sum_all=np.sum(gfg)
print("sum of all elements",sum_all)
#ii. Sum of all elements row-wise
row_sum=np.sum(gfg,axis=1)
print("row sum",row_sum)
# iii. Sum of all elements column-wise
col_sum=np.sum(gfg,axis=0)
print("column sum",col_sum)
```

```python
'''Q.3 You are given a weekly sales dataset and need to perform various operations
using NumPy broadcasting.
a) Generate your unique sales dataset:'''

# Take the sum of the ASCII values of the initials of your first and lastname. Call this value X. (If your initials are A B → ASCII sum = 65
#+ 66 = 131 → sales = [131, 181, 231, 281, 331].)
x=ord('Y')+ord('S')
sales=np.array([x,x+50,x+100,x+200])
print("sales are",sales)
# Create a NumPy array sales with values [X, X+50, X+100, X+150,
#X+200]. '''
sales_1=np.array([x,x+50,x+100,x+150,x+200])
print("sales as",sales_1)
'''b) Compute your personalized tax rate as ((X % 5) + 5) / 100. Use broadcasting to apply this tax rate to each sales value.'''
sales=np.array([172,222,272,322,372])
tax_rate=((172%5)+5)/100
tax_amount=sales*tax_rate
print("tax amount",tax_amount)
'''c) Adjust sales based on discount:'''
# If sales < X+100, apply a 5% discount.
sales=np.array([172,222,272,322,372])
discount=np.where(sales<272,sales*0.95,sales)
print("discount",discount)
#  If sales >= X+100, apply a 10% discount. '''
sales=np.array([172,222,272,322,372])
discount=np.where(sales>272,sales*0.9,sales)
print("discount as",discount)
'''d) Expand sales data for multiple weeks:'''
# Create a 3×5 matrix representing three weeks of sales by stacking sales three times using broadcasting.
sales=np.array([172,222,272,322,372])
weekly_sales=np.tile(sales,(3,1))
print("weekly sales as\n",weekly_sales)
# Increase sales by 2% per week using element-wise broadcasting.
sales=np.array([172,222,272,322,372])
weekly_sales=np.tile(sales,(3,1))
week_multipliers=np.array([1.00,1.02,1.04]).reshape(3,1)
adjusted_sales=weekly_sales*week_multipliers
print("adjusted sales as \n",adjusted_sales)
```

```python
'''Q4. Generate x values using np.linspace() from -10 to 10 with 100 points. Use
each function from the list below and compute y values using NumPy:
 Y = x2
 Y = sin(x)
 Y = ex
 Y = log(|x| + 1)
Plot the chosen function using Matplotlib. Add title, labels, and grid for clarity. '''
import matplotlib.pyplot as plt
x=np.linspace(-10,10,100)
y1=x**2
y2=np.sin(x)
y3=np.exp(x)
y4=np.log(np.abs(x)+1)
# Plot the functions
plt.figure(figsize=(10, 6))
plt.plot(x, y1, label=r'$y = x^2$', color='b')
plt.plot(x, y2, label=r'$y = \sin(x)$', color='r')
plt.plot(x, y3, label=r'$y = e^x$', color='g')
plt.plot(x, y4, label=r'$y = \log(|x| + 1)$', color='m')

# Add labels, title, and grid
plt.xlabel("x values")
plt.ylabel("y values")
plt.title("Graphs of Different Functions")
plt.legend()  # Show legend
plt.grid(True)  # Add grid
```
